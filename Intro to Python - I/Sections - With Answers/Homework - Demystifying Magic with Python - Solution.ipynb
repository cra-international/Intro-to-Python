```python
import random
```

### Part 2: Define `add_digits()`

Next define a function that takes a number as an input and returns the sum of the digits. We'll need this later!

```python
def add_digits(num):
    total = 0
    str_num = str(num)
    for digit in str_num:
        total += int(digit)
    return total
```

### Part 3: Simulate the Trick

Next, write some Python code that simulates Dan White's magic trick. The below code should 
1. Simulate the trick 100,000 times
2. For each simulation, first pick a  random number between 2 and 9. (We ignore 1 here because multiplying by 1 doesn't do anything)
3. Once a number is picked, the code should keep by *another* random number between 2 and 9 until the product is over 1,000,000
4. Once the product is greater than or equal to 1,000,000, the program should get the sum of the digits of that number by calling the `add_digits()` function defined above
5. Finally, the code should store the sum of the digits of your number somewhere for later reference

```python
num_list = []
for i in range(100000):
    total = 1
    while total < 1000000:
        rand_digit = random.randint(2, 9)
        total *= rand_digit
    digit_sum = add_digits(total)
    num_list.append(digit_sum)
```

### Part 4: Compile the Results

Finally, for each number 1 through 100 (since these are all the numbers on Dan White's magic sheet), count how many times your simulation produced that number. Then, store that count in a dictionary where the *key* is the magic number and the *value* is how many times that magic number appeared in your simulation.

```python
counts = {}
for i in range(1, 101):
    num_count = num_list.count(i)
    counts[i] = num_count
```

### Part 5: Explore the Results

And that's it! Look at your dictionary. Do all values between 1 and 100 appear equally as often? If not, which one appear most frequently?

We'll go over the answers at the beginning of Python Part 2.

```python
print(counts)
```
